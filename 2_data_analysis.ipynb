```python
plt.figure(figsize= (6,6))
plt.pie(x= df.book_table.value_counts(), labels= df.book_table.value_counts().index, colors= ['blue', 'yellow'], autopct= '%.0f%%', explode= (0, 0.2), shadow= True, startangle= 130)

plt.title("Book Table Distribution")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (12, 8))
sns.set_style('darkgrid')
sns.barplot(data= df, x= "book_table", y= "rate", order= df.book_table.value_counts().index, label= "Book Table")

plt.grid(True, linewidth= 0.8, linestyle= "--", color= "grey", alpha= 0.3)

plt.title("Book Table V/S Rate", fontsize= 16, fontweight= 'bold')
plt.xlabel("Book Table", fontsize= 12)
plt.ylabel("Rate", fontsize= 12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Set style and context
sns.set_style("whitegrid")  # whitegrid is better for controlling gridlines

# Create the plot and get the axes
ax = sns.barplot(data=df, x='book_table', y='rate', palette='Set2')

# Make sure only horizontal gridlines are shown
ax.grid(True, axis='y', linestyle='--', linewidth=0.7, alpha=0.5)
ax.grid(False, axis='x')  # Remove vertical gridlines

# Titles and labels
ax.set_title("Book Table vs. Rate", fontsize=14, fontweight='bold', color='#FF8C00')
ax.set_xlabel("Book Table")
ax.set_ylabel("Rate")

plt.tight_layout()
plt.show()
```

```python
x= df.groupby('location')[['rate', 'votes']].mean()
x.sort_values(by= ['rate', 'votes'], ascending= False, inplace= True)
x= x.head(10)
l= x.index.to_list()
```

```python
x
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.barplot(data= x, x= x.index, y= x.rate, palette='Set2', ax=ax, hue= x.index)
ax.grid(True, axis='y', linestyle='--', linewidth=0.7, alpha=0.5)
ax.set_title('Best Location based on Rating')
ax.set_xlabel('Location')
ax.set_ylabel('Rating')
ax.axhline(0, color='black', linewidth=1); ax.axvline(-0.5, color='black', linewidth=1)
ax.legend()
fig.tight_layout()
plt.show()
```

```python
sns.barplot(x=x.index,y=x.rate,data=x,palette='magma')
plt.xticks(rotation=90)
plt.title("Best location by rating")
plt.show()
```

```python
plt.figure(figsize= (12, 8))
sns.barplot(x= x.index, y= x.rate, linewidth= 2, linestyle= ":", palette= 'rocket')

plt.grid(True, linewidth= 1, linestyle= ":", color= "grey", alpha= 0.6)

plt.title("Best Location Based on Rating", fontsize= 16, fontweight= "bold")
plt.xlabel("Location", fontsize= 12)
plt.ylabel("Rating", fontsize= 12)
plt.xticks(rotation= 90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize= (12, 8))
sns.barplot(x= x.index, y= x.votes, palette= 'viridis', linewidth= 1.5, linestyle= ":", edgecolor= "black", width= 0.8)

plt.grid(True, linewidth= 1, linestyle= ":", alpha= 0.6)

plt.title("Votes in Location with Best Rating", fontsize= 16, fontweight= 'bold')
plt.xlabel("Location", fontsize= 12)
plt.ylabel("Votes", fontsize= 12)
plt.xticks(rotation= 90)
plt.tight_layout()
plt.show()
```

```python
df['rest_type_list'] = df['rest_type'].str.split(',\s*')  # split on comma and optional space
df_exploded = df.explode('rest_type_list')
rest_type_rating = df_exploded.groupby('rest_type_list')['rate'].mean().sort_values(ascending=False)
print(rest_type_rating)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=rest_type_rating.values, y=rest_type_rating.index, palette='viridis')
plt.xlabel('Average Rating')
plt.ylabel('Rest Type')
plt.title('Average Rating by Individual Rest Type')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
